    ATTRITION RATE BY GENDER FOR DIFFERENT AGE GROUP

```python
gender_data = my_data[['Attrition','CF_age band','Gender']]
gender_data.head(5)
```

```python
groupby_gender = gender_data.groupby(["CF_age band","Gender"], as_index = False)['Attrition'].count()
groupby_gender
```

```python
def percentage(group_by):
    number = group_by.Attrition.astype(float)
    group_by['percent'] = round(number*100/number.sum(),2)
    return group_by
group_bygender = percentage(groupby_gender)
group_bygender
```

```python
data_under25 = group_bygender[(group_bygender['CF_age band'] == "Under 25")]
data_25_34 = group_bygender[(group_bygender['CF_age band'] == "25 - 34")]
data_35_44 = group_bygender[(group_bygender['CF_age band'] == "35 - 44")]
data_45_54 = group_bygender[(group_bygender['CF_age band'] == "45 - 54")]
data_Over55 = group_bygender[(group_bygender['CF_age band'] == "Over 55")]
```

```python
fig, ax = plt.subplots(figsize = (12,4))

#UNDER25
x = data_under25
plt.title('ATTRITION RATE BY GENDER FOR DIFFERENT AGE UNDER 25')
plt.pie(x["percent"],
        labels = (x["percent"]), 
        colors = ["red","blue"],
        autopct = '%1.2f%%',
        shadow = False,
        startangle=90)
plt.legend(x["Gender"],loc = "upper left")
plt.show()

#25-34
y = data_25_34
plt.title('ATTRITION RATE BY GENDER FOR DIFFERENT AGE 25-34')
plt.pie(y["percent"],
        labels = y["percent"], 
        colors = ["red","blue"],
        autopct = '%1.2f%%',
        shadow = False,
        startangle=90)
plt.legend(y["Gender"],loc = "upper left")
plt.show()

#35 - 44
z = data_35_44
plt.title('ATTRITION RATE BY GENDER FOR DIFFERENT AGE 35-44')
plt.pie(z["percent"],
        labels = z["percent"], 
        colors = ["red","blue"],
        autopct = '%1.2f%%',
        shadow = False,
        startangle=90)
plt.legend(z["Gender"],loc = "upper left")
plt.show()

#45 - 54
t = data_45_54
plt.title('ATTRITION RATE BY GENDER FOR DIFFERENT AGE 45-54')
plt.pie(t["percent"],
        labels = t["percent"], 
        colors = ["red","blue"],
        autopct = '%1.2f%%',
        shadow = False,
        startangle=90)
plt.legend(t["Gender"],loc = "upper left")
plt.show()

#Over 55
r = data_Over55
plt.title('ATTRITION RATE BY GENDER FOR DIFFERENT AGE OVER 55')
plt.pie(r["percent"],
        labels = r["percent"], 
        colors = ["red","blue"],
        autopct = '%1.2f%%',
        shadow = False,
        startangle=90)
plt.legend(r["Gender"],loc = "upper left")
plt.show()
```
